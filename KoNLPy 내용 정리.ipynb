## 4. 육아휴직 관련 법안에 대한 NLP

```python
import nltk
```

```python
from konlpy.corpus import kobill
files_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read() # 육아휴직 관련 법안 제 1809890호
doc_ko
```

```python
# 트위터 엔진
from konlpy.tag import Okt
t = Okt()
tokens_ko = t.nouns(doc_ko) # 명사 분석
tokens_ko
```

```python
# 수집된 단어의 횟수, 고유한 횟수
ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
print(len(ko.tokens))
print(len(set(ko.tokens)))
ko.vocab()
```

```python
plt.figure(figsize=(20,10))
ko.plot(50)
plt.show()
```

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', 'x', 
              '의', '자', '에', '안', '번', '호', '을', '이', '다', '만', '로', '가', '를']
ko = [each_word for each_word in ko if each_word not in stop_words]
```

```python
ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
plt.figure(figsize=(20,10))
ko.plot(50)
plt.show()
```

```python
ko.count('초등학교')
```

```python
plt.figure(figsize=(20,10))
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])
```

```python
# 원하는 단어의 문서 내 위치를 개략적으로 파악
ko.concordance('초등학교')
```

```python
# 연어 파악
ko.collocations()
```

```python
data = ko.vocab().most_common(150)
wordcloud = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf', relative_scaling=0.2,
                     background_color='white').generate_from_frequencies(dict(data))
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

## 5. Naive Bayes Classfier
- 지도학습(supervised learning)의 한 종류
- 두 사건을 서로 독립이라고 가정하고 각각의 조건부확률을 활용
- 참고1 : https://gomguard.tistory.com/69
- 참고2 : https://ratsgo.github.io/machine%20learning/2017/05/18/naive/


### 1. 영문

```python
from nltk.tokenize import word_tokenize
import nltk
```

```python
train = [('i like you', 'pos'), ('i hate you', 'neg'), ('you like me', 'neg'), ('i like her', 'pos')]
```

```python
# train 문장에서 사용된 전체 단어 (말뭉치)
all_words = set(word.lower() for sentence in train for word in word_tokenize(sentence[0]))
all_words
```

```python
# 말뭉치를 기준으로 train 문장에 속한 단어인지 아닌지 기록
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1]) for x in train]
t
```

```python
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
# 'hate'가 없을 경우 긍정일 비율이 1.7:1
```

```python
test_sentence = 'i like MeRui'
test_sent_features = {word.lower(): (word in word_tokenize(test_sentence.lower())) for word in all_words}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```

### 2. 한글

```python
from konlpy.tag import Okt
pos_tagger = Okt()
```

```python
train = [('메리가 좋아', 'pos'), ('고양이도 좋아', 'pos'), ('난 수업이 지루해', 'neg'), 
         ('메리는 예쁜 고양이야', 'pos'), ('난 마치고 메리랑 놀거야', 'pos')] 
```

```python
all_words = set(word.lower() for sentence in train for word in word_tokenize(sentence[0]))
all_words
```

```python
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1]) for x in train]
t
```

```python
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

```python
test_sentence = '난 수업을 마치면 메리랑 놀거야'
test_sent_features = {word.lower(): (word in word_tokenize(test_sentence.lower())) for word in all_words}
test_sent_features
```

```python
classifier.classify(test_sent_features) # 결과가 잘못 나올 수 있으므로 한글에 대해서는 형태소 분석이 필요함
```

```python
# 형태소 분석
def tokenize(doc):
    return ['/'.join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)]
```

```python
train_docs = [(tokenize(row[0]), row[1]) for row in train]
train_docs
```

```python
# 전체 말뭉치
tokens = [t for d in train_docs for t in d[0]]
tokens
```

```python
# 말뭉치에 있는 단어가 있는지 아닌지를 구분하는 함수
def term_exists(doc):
    return {word: (word in set(doc)) for word in tokens}
```

```python
train_xy = [(term_exists(d), c) for d, c in train_docs]
train_xy
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_xy)
```

```python
test_sentence = '난 수업을 마치면 메리랑 놀거야'
test_docs = pos_tagger.pos(test_sentence)
test_docs
```

```python
test_sent_features = {word: (word in tokens) for word in test_docs}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```
